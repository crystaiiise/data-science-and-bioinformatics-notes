```python
# Source: https://time.geekbang.org/column/intro/100021701?tab=intro

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 数据准备
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)
# Matplotlib画散点图
plt.scatter(x, y,marker='x') # marker代表了散点的符号，比如“x”、“>”或者“o”
plt.show()
# Seaborn画散点图
df = pd.DataFrame({'x': x, 'y': y})
sns.jointplot(x="x", y="y", data=df, kind='scatter');  # **需要传入DataFrame类型**
plt.show()
```

```python
# 数据准备
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]
# Matplotlib画折线图
plt.plot(x, y) # 直接就是plot()
plt.show()
# Seaborn画折线图
df = pd.DataFrame({'x': x, 'y': y})
sns.lineplot(x="x", y="y", data=df)
plt.show()
```

```python
a = np.random.randn(100) 
s = pd.Series(a) 
# Matplotlib画直方图 
# 直方图看**连续数据**出现的频次（需将横坐标分成一定数量的区间，称“箱子”），柱图看的是离散数据。
# plt.hist()和sns.histplot()都默认10个箱子
plt.hist(s)  
plt.show()
# Seaborn画直方图
sns.histplot(s, kde=False) 
# 'kde': Kernel Density Estimation, creates a smooth line curve that helps visualize the overall trends.
plt.show()
sns.histplot(s, kde=True)
plt.show()
```

```python
# 数据准备, 使用Seaborn中自带的数据集flights
flights = sns.load_dataset("flights")
print(flights.head())
data=flights.pivot(index='year', columns='month', values='passengers')
# pivot/pivot_table提供整理DataFrame的方式，原本df当中各变量的数据存放方式全是按列向量（用'_'索引嘛）
# 整理之后index是横轴想看的变量，columns是纵轴想看的。pivot_table还加上了一个aggfunc参数，

# Seaborn画热力图
sns.heatmap(data)
plt.show()
```

```python
# 蜘蛛图/radar plot用于显示一对多关系。
from matplotlib.font_manager import FontProperties  
# 数据准备
labels=np.array(["A","B","C","D","E","F"])
stats=[83, 61, 95, 67, 76, 88]

# 画图数据准备，角度、状态值
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
stats=np.concatenate((stats,[stats[0]]))
# Ensuring the radar plot forms a closed loop when plotted.
angles=np.concatenate((angles,[angles[0]]))
labels=np.concatenate((labels,[labels[0]]))

# Matplotlib画蜘蛛图
fig = plt.figure()
# Adds a subplot (1 row, 1 col, only 1 subplot) with a polar coordinate system.
ax = fig.add_subplot(111, polar=True)   
# Plots the stats data against the angles using a line with circular markers ('o-').
ax.plot(angles, stats, 'o-', linewidth=2)
# Fills the area under the radar plot to visually emphasize the data. 
ax.fill(angles, stats, alpha=0.25)
# Converts the angles from radians to degrees and sets grid labels around the polar plot using the labels
ax.set_thetagrids(angles * 180/np.pi, labels) 
plt.show()
```

```python
tips = sns.load_dataset("tips")
print(tips.head(10))

# 用Seaborn画二元变量分布图（散点图，核密度图，Hexbin图）
sns.jointplot(x="total_bill", y="tip", data=tips, kind='scatter')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='kde')
sns.jointplot(x="total_bill", y="tip", data=tips, kind='hex')
sns.jointplot(x="size", y="tip", data=tips, kind='hex')

plt.show()
```

```python
iris = sns.load_dataset('iris')
# 用Seaborn展示成对关系（主对角线上能看到每个变量自身分布情况）
# pairplot就像在DataFrame中使用describe()统计函数一样方便
sns.pairplot(iris)
plt.show()
```
